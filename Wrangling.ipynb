```python
data = pd.read_csv('/content/wrangling/data/sharks.csv', low_memory=False)
type = data['Type']
# print(type.value_counts())
type = type.replace(['Sea Disaster', 'Boat', 'Boating', 'Boatomg'], 'Watercraft')
type = type.replace(['Invalid', 'Questionable', 'Unconfirmed', 'Unverified', 'Under investigation'], 'Unclear')
print(type.value_counts())
data['Type'] = type
```

Here I cleaned the Type variable. I noticed that a lot of types documented were essentially synonyms, and therefore there were more categories in this variable than were necessary. I combined all water/boat-related types into the "Watercraft" category, and I combined all unknown/questionable types into an "Unclear" category. I then put the cleaned variable back into the data set.

*Part 3*

```python
import numpy as np

url = 'http://www.vcsc.virginia.gov/pretrialdataproject/October%202017%20Cohort_Virginia%20Pretrial%20Data%20Project_Deidentified%20FINAL%20Update_10272021.csv'
data = pd.read_csv(url,low_memory=False)
released = data['WhetherDefendantWasReleasedPretrial']
# print(released.value_counts())
released = released.replace(9,np.nan)
print(released.value_counts())
data['WhetherDefendantWasReleasedPretrial'] = released
```

In the codebook for this data set, we can find that 1 corresponds to 'Released during pre-trial period', 0 corresponds to 'Detained entire pre-trial period', and 9 corresponds to 'Unknown'. So, here I replaced all of the 9s with np.nan since they are missing values, and then put the cleaned variable back into the data set.

*Part 4*

```python
sentence = data['ImposedSentenceAllChargeInContactEvent']
s_type = data['SentenceTypeAllChargesAtConvictionInContactEvent']
# print(sentence.value_counts())
# print(s_type.value_counts())
sentence = pd.to_numeric(sentence, errors='coerce')
print('Total missing data points before cleaning: ', sum(pd.isnull(sentence)))
sentence = sentence.mask(s_type == 4, 0)
sentence = sentence.mask(s_type == 9, np.nan)
print('Total missing data points: ', sum(pd.isnull(sentence)))
data['ImposedSentenceAllChargeInContactEvent'] = sentence
```

Here I cleaned the ImposedSentenceAllChargeInContactEvent variable. From the codebook, you can see that there are actually only 3 types of sentences, corresponding to 0, 1, and 2 in the SentenceTypeAllChargesAtConvictionInContactEvent variable. So, type '4' in that variable should not have any events, and we already know that '9' corresponds to unknown data. So, I changed the 4s to be 0, and the 9s to be np.nan. This brought my amount of missing data points from 9053 to just 274. I then put the cleaned variable back into the data set.
